````python
## Part 3: Handling Missing Values

### Task 3.1: Create a Cleaning Strategy
"""
Before handling missing values, create a strategy for each column. Fill in the table:

in the table:

| Column | Strategy | Justification |
|--------|----------|---------------|
| customer_id | | |
| customer_name | | |
| email | | |
| phone | | |
| registration_date | | |
| age | | |
| purchase_amount | | |
| product_category | | |
| payment_method | | |
"""

# **Common Strategies:**
# - Drop rows with missing values
# - Fill with mean/median/mode
# - Fill with a constant value
# - Forward fill / Backward fill
# - Fill with interpolation


#  Part 3: Handling Missing Values
## Task 3.1: Create a Cleaning Strategy
"""
| Column | Strategy | Justification |
|--------|----------|---------------|
| customer_id |drop row |id is primary key to get customers data |
| customer_name |unknown |names can be unique |
| email |drop rows |email is unique, so missing email reduce usability |
| phone |not provided | phone number is optional and not always available|
| registration_date |drop rows |date is importent to analyse time |
| age |median |Median handles outliers better than mean |
| purchase_amount |median | |
| product_category |mode |frequent category |
| payment_method |mode| payment is frequent use|
"""
````

````python
### Task 3.2: Implement Missing Value Handling

# Create a copy of the original DataFrame called `df_clean` and implement your strategy:

# Create a copy
# df_clean = df.copy()

# Implement your strategy for each column
# Your code here

# -----------------------------------------
# Task 3.2: Implement Missing Value Handling
# -----------------------------------------
print("="*60)
print("Task 3.2: Implementing Missing Value Handling")
print("="*60)

# copy of the original DataFrame
df_clean = df.copy()

# 1. droping rows with missing customer id's
# used gpt to understand subset
# used subnet to drop missing values from specific columns
df_clean = df_clean.dropna(subset=["customer_id"])

# 2. Filling the  missing customer_name
df_clean["customer_name"] = df_clean["customer_name"].fillna("Unknown")

# 3. Drop rows with missing email
df_clean = df_clean.dropna(subset=["email"])

# 4. Filling missing phone
df_clean["phone"] = df_clean["phone"].fillna("Not Provided")

# 5. Droping rows with missing registration_date
df_clean = df_clean.dropna(subset= ["registration_date"])

# 6. Filling missing age with median
df_clean["age"] = df_clean["age"].fillna(df_clean["age"].median())

# 7. Fill missing purchase_amount with median
df_clean["purchase_amount"] = df_clean["purchase_amount"].fillna(
    df_clean["purchase_amount"].median()
)

# 8. Fill missing product_category with mode
# mode return series
# and [0] selects the first most frequent value from the series
df_clean["product_category"] = df_clean["product_category"].fillna(
    df_clean["product_category"].mode()[0])
# if missing value again found
df_clean["payment_method"] = df_clean["payment_method"].fillna("unknown")

# 9. Fill missing payment_method with mode
df_clean["payment_method"] = df_clean["payment_method"].fillna(
    df_clean["payment_method"].mode([0]))

print("\n--- Missing Value Handling completed successfully ---\n")
print(df_clean)
````

````python
## Part 7: Handling Outliers and Invalid Values

### Task 7.1: Fix Invalid Ages
"""
Identify and fix:
- Negative ages
- Ages greater than 120

**Strategy:** Replace invalid ages with the median age.
"""
# Your code here

print("="*60)
print("Part 7: Handling Outliers and Invalid Values")
print("="*60)
# -----------------------------
### Task 7.1: Fix Invalid Age
# -----------------------------

print("\n--- Task 7.1: Fix Invalid Age ---")

# calculating median age
median_age = df_clean["age"].median()

# replacing negative age and age > 120 with median age
# find rows with invalid ages
# and loc update only those rows
df_clean.loc[(df_clean["age"]< 1) | (df_clean["age"]>120), "age"] = median_age

print("\nInvalid age fixed successfully.")
print(df_clean["age"])
````

````python
### Task 7.2: Handle Purchase Amount Outliers
"""
Use the IQR (Interquartile Range) method to:
1. Calculate Q1, Q3, and IQR
2. Determine lower and upper bounds
3. Identify outliers
4. Handle outliers by capping them at the bounds

**Formula:**
- IQR = Q3 - Q1
- Lower Bound = Q1 - 1.5 × IQR
- Upper Bound = Q3 + 1.5 × IQR

```python
"""
# Your code here

# --------------------------------------------
# Task 7.2: Handle Purchase Amount Outliers
# --------------------------------------------
print("\n--- Task 7.2: Handle Purchase Amount Outliers ---")

# Purchase amounts cannot be negative
df_clean.loc[df_clean["purchase_amount"] < 0, "purchase_amount"] = 0

# 1. Calculate Q1, Q3, and IQR
print("\n--- 1. Calculate Q1, Q3, and IQR ---")
# q1 -> 25% of data lies below this value
q1 = df_clean["purchase_amount"].quantile(0.25)
#  q3 -> 75% of data lies below this value
q3 = df_clean["purchase_amount"].quantile(0.75)

# calculating Interquartile Range(IQR)
iqr = q3 - q1

# -----------------------------------
# 2. Determine lower and upper bounds
# -----------------------------------
print("\n--- 2. Determine lower and upper bounds ---")

# calculating lower bound
lower_bound = q1 - 1.5 * iqr

# calculating upper bound
upper_bound = q3 + 1.5 * iqr

# -----------------------------------
# 3. Identify outliers
# -----------------------------------
print("\n--- 3. Identify outliers ---")

outliers = df_clean[
    (df_clean["purchase_amount"] < lower_bound) |
    (df_clean["purchase_amount"] > upper_bound)
]
# ---------------------------------------------
# 4. Handle outliers by capping them at the bounds
# ---------------------------------------------
print("\n--- 4. Handle outliers by capping them at the bounds ---")

# Cap values below the lower bound to the lower bound
df_clean.loc[df_clean["purchase_amount"] < lower_bound, "purchase_amount"] = lower_bound

# Cap values above the upper bound to the upper bound
df_clean.loc[df_clean["purchase_amount"] > upper_bound, "purchase_amount"] = upper_bound

# **Questions:**
# - How many outliers were identified?
# - What were the outlier values?
# - What is the new range of purchase amounts?

print("\n--- Questions ---")
# number of outliers
print("\n1. How many outliers were identified?")
print(outliers.shape[0])

# values of outliers
print("\n2. What were the outlier values?")
print(outliers["purchase_amount"].values)

# new range of purchase amount
# used gpt to understand how to calculat new range
new_min = df_clean["purchase_amount"].min()
new_max = df_clean["purchase_amount"].max()
print("\n3. What is the new range of purchase amounts?")
print(f"{new_min} to {new_max}")
````
