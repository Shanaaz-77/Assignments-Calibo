````python
## Part 4: Outlier Detection (20 points)

### Task 4.1: Implement IQR Method
# Write a function to detect outliers using the IQR (Interquartile Range) method:
# - Calculate Q1 (25th percentile) and Q3 (75th percentile)
# - Calculate IQR = Q3 - Q1
# - Define outliers as values < Q1 - 1.5×IQR or > Q3 + 1.5×IQR

# ```python
# def detect_outliers_iqr(data, column):

# Your code here

# ---------------------------------------
# Task 4.1: Implement IQR Method
# ---------------------------------------
print("\n--- Task 4.1: Implement IQR Method ---")

#  Creating a function to detect outliers
def detect_outliers_iqr(data, column):
    # calculate Q1 and Q3
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)

    # calculating IQR
    IQR = Q3 - Q1

    # defining boundaries
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # detecting outliers
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
````

````python
### Task 4.2: Detect Outliers
# Apply your function to detect outliers in:
# - quantity
# - price_per_unit
# - total_amount

# Report the number of outliers in each column.

# Your code here

# ---------------------------------------
# Task 4.2: Detect Outliers
# ---------------------------------------
print("\n--- Task 4.2: Detect Outliers ---")

# given colunms to check
cols_to_check = ["quantity", "price_per_unit", "total_amount"]

# detecting outliers
for col in cols_to_check:
  outliers = detect_outliers_iqr(df, col)
  print(f"\n--- Number of outliers in  {col} ---")
  print(len(outliers))

````

````python
### Task 5.3: Test for Normality
# Use the Shapiro-Wilk test to formally test if the distributions are normal. (p-value > 0.05 suggests normal distribution)

# Your code here

# ---------------------------------------
# Task 5.3: Test for Normality
# ---------------------------------------
print("="*60)
print("Task 5.3: Test for Normality")
print("="*60)

from scipy.stats import shapiro

# selecting numerical columns
num_cols = ['age', 'quantity', 'price_per_unit', 'total_amount']

for col in cols:
    stat, p = shapiro(df[col])
    # Used gpt to understand Shapiro-Wilk test
    print(f"\n--- Shapiro-Wilk Test for {col} ---")
    print(f"Statistic = {stat:.4f}, p-value = {p:.4f}")

    # interpretation
    # 0.05 is the default significance level
    if p > 0.05:
        # Null hypothesis (H0) for Shapiro-Wilk
        print("=> Distribution is normal (fail to reject H0)")
    else:
        print("=> Distribution is not normal (reject H0)")

````
