````python
## Exercise 6: Real-World Application - Student Grade Analysis (25 points)

"""You are given grade data for 5 students across 4 subjects (Math, Science, English, History):

```python

grades = np.array([
    [85, 90, 78, 88],
    [92, 88, 95, 90],
    [76, 82, 80, 75],
    [88, 85, 90, 92],
    [95, 92, 88, 94]
])
"""
"""
**Tasks:**

1. **Calculate the average grade for each student** (5 points)
   - Output should show Student 1: X.XX, Student 2: Y.YY, etc.

2. **Calculate the average grade for each subject** (5 points)
   - Output: Math: X.XX, Science: Y.YY, etc.

3. **Find the highest and lowest grades in each subject** (5 points)
   - Output for each subject: Highest = X, Lowest = Y

4. **Identify students with an average grade above 85** (3 points)
   - Print student number and their average

5. **Calculate the overall class average** (2 points)
   - Average of all grades across all students

6. **Normalize all grades to a scale of 0-10** (5 points)
   - Use min-max normalization formula: (value - min) / (max - min) * 10
   - Print the normalized grades array

**Bonus:** Identify the top-performing student (+2 points)

**Your Code:**
python
"""      


## Exercise 6: Real-World Application - Student Grade Analysis
print("="*60)
print("Exercise 6: Student Grade Analysis")
print("="*60)

# Given matrix
grades = np.array([
    [85, 90, 78, 88],
    [92, 88, 95, 90],
    [76, 82, 80, 75],
    [88, 85, 90, 92],
    [95, 92, 88, 94]
])

# There are 4 subjects
subject_name = ["Maths", "Science", "English", "History"]
print("\nStudent grades:\n",grades)
print("shape:", grades.shape)

# 1. average grade for each student(axis = 1)
print("\n--- Task 1: Student Averages ---")
student_avg = np.mean(grades, axis = 1) #each student marks are row wise

# used gpt to print output in required format
for student_num, avg in enumerate(student_avg, start=1):
    # +1 converts 0-based index to student number
    print(f"Student {student_num}: {avg:.2f}")

 # 2. average grade for each subject(axis = 0)
print("\n--- Task 2: Subject Averages ---")
subject_avg = np.mean(grades, axis = 0) #each subject average, column wise

# used gpt to print output in required format
for subject, avg in zip(subject_name, subject_avg):
    print(f"{subject}: {avg:.2f}")

# 3. highest and lowest grades in each subject(axis = 0)
print("\n--- Task 3: Highest and Lowest per subject ---")
subject_max = np.max(grades, axis = 0) # highest grade in subjects
subject_min = np.min(grades, axis = 0) # lowest grade in subject

# Loop to display reslut in required format
for subject, high, low in zip(subject_name, subject_max, subject_min):
    print(f"{subject}: Highest = {high}, Lowest = {low}")

#Students with average above 85 
print("\n--- Task 4 : Students with average above 85 ---")

above_85_avg = student_avg > 85
for i, avg in enumerate(student_avg[above_85_avg], start=1):
    # used gpt to display required output
    # np.where(above_85_mask)[0] gives indices of True values
    # i-1 is used because enumerate starts from 1
    student_num = np.where(above_85_avg)[0][i-1] + 1
    print(f"Student {student_num}: {avg:.2f}")

# overall class average
class_avg = np.mean(grades)
print("\n--- Task 5 : overall class average ---")
print("Overall average:",class_avg)

# Normalizing all grades to a scale of 0-10
# minimum of grades
min_grade = np.min(grades)
#  maximum of grades
max_grade = np.max(grades)

# Using min-max normalization formula
normalized_grades = (grades - min_grade) / (max_grade - min_grade) * 10

print("\n--- Task 6 : normalized grades array ---")
print(np.round(normalized_grades, 2))

# Identifying the top-performing student
print("\n--- bonus : top-performing student ---")
top_student = np.argmax(student_avg) + 1
print("Student",top_student)



"""
**Expected Output Format:**
```
--- Task 1: Student Averages ---
Student 1: XX.XX
...

--- Task 2: Subject Averages ---
Math: XX.XX
...

(Continue for all tasks)

"""
````

````python
## Exercise 7: Challenge - Sales Data Analysis (30 points)

"""You are analyzing sales data for a company with 4 products over 6 months:

python
sales_data = np.array([
    [120, 135, 150, 145, 160, 175],  # Product A
    [90, 95, 100, 110, 115, 120],    # Product B
    [150, 145, 155, 160, 170, 180],  # Product C
    [80, 85, 82, 88, 92, 95]         # Product D
])
**Tasks:**

1. **Calculate total sales for each product** (5 points)
   - Sum across all months for each product

2. **Calculate total sales for each month** (5 points)
   - Sum across all products for each month (Jan to Jun)

3. **Identify the month with the highest total sales** (4 points)
   - Print the month name and total sales

4. **Calculate the growth rate for each product** (6 points)
   - Growth rate = ((Last month - First month) / First month) × 100
   - Print as percentage for each product

5. **Find products with average monthly sales greater than 120k** (4 points)
   - Calculate average for each product
   - Print products meeting the criteria

6. **Calculate quarter-wise sales** (6 points)
   - Q1: January-March
   - Q2: April-June
   - Print sales for each product in each quarter

**Bonus:** Predict next month (July) sales for each product (+5 points)
- Use a simple method: average of last 2 months + average monthly growth

**Your Code:**
python
"""

## Exercise 7: Challenge - Sales Data Analysis
print("="*50)
print("Exercise 7: Challenge - Sales Data Analysis")
print("="*50)

# sales data(rows = products, columns = months) 
sales_data = np.array([
    [120, 135, 150, 145, 160, 175], # product A
    [90, 95, 100, 110, 115, 120], # product B
    [150, 145, 155, 160, 170, 180], # product C
    [80, 85, 82, 88, 92, 95] # product D
])

# Product names in the sales data          
product_names = np.array(["Product A", "Product B", "Product C", "Product D"])
# Month names in the sales data             
month_names = np.array(["Jan", "Feb","Mar", "Apr", "May", "Jun"])

#---------------------------------
# 1. Total sales for each product 
#---------------------------------
product_total_sales = np.sum(sales_data, axis = 1) # (axis = 1, row wise)
print("\n--- Task 1 : Total sales for each product ---")
# Used gpt to know how to use zip 
# used zip to pair each product with its total sales
# and display them together
for product, total in zip(product_names, product_total_sales):
    print(f"{product}: {total}")
#---------------------------------
# 2. Total sales for each month
#---------------------------------
monthly_total_sales = np.sum(sales_data, axis = 0) # (axis = 0, column wise)
                   
print("\n--- Task 2 : Total sales for each month ---")
# use zip to pair each month with its total sales
# and display them together
for month, total in zip(month_names, monthly_total_sales):
    print(f"{month}: {total}")

#--------------------------------------
# 3. Month with the highest total sales
#--------------------------------------
print("\n--- Task 3 : Month with the highest total sales ---")
# usign argmax to get highest index value
highest_month = np.argmax(monthly_total_sales)
    
print("\nHighest sales month:", month_names[highest_month])
print("with sales", monthly_total_sales[highest_month])

#------------------------------------
# 4. Growth rate for each product
#------------------------------------

first_month_sales = sales_data[:, 0] # january sales
last_month_sales = sales_data[:, -1] # june sales

# using Growth rate formula
growth_rate = ((last_month_sales - first_month_sales) / first_month_sales ) * 100

print("\n--- Task 4 : Growth rate for each product ---")
# Use zip to pair each product name with its growth rate
# and display them together
for product, growth in zip(product_names, growth_rate):
    print(f"{product}: {growth:.2f}%")

#---------------------------------------------------
# 5. Products with average monthly sales > 120k 
#---------------------------------------------------
# Calculating average monthly sale for each product
product_avg_sales =np.mean(sales_data, axis = 1) # row wise
# checking which products have avg sale > 120k
above_120_avg = product_avg_sales > 120 # condition

print("\n--- Task 5 : Products with average monthly sales > 120k ---")
# use zip to display only products with avg sale >120k
for product, avg in zip(product_names[above_120_avg], growth_rate[above_120_avg]):
    print(f"{product}: {avg:.2f}%")

#--------------------------------------------------
# 6. Quarter-wise sales
#--------------------------------------------------

# Q1: January-March(columns 0 to 2)
# used gpt to get Q1
q1_sales = np.sum(sales_data[:, 0:3], axis = 1)
# Q2: April to June (columns 3 to 5)
q2_sales = np.sum(sales_data[:, 3:6], axis = 1)

print("\n--- Task 6 : Quarter-wise sales ---")

# use zip to display Q1 and Q2 sales for each product
for product, q1, q2 in zip(product_names, q1_sales, q2_sales):
    print(f"{product} -> Q1: {q1}, Q2 : {q2}")
    
#-------------------------------------------------------
# Bonus:Predict next month (July) sales for each product
#-------------------------------------------------------
# used gpt to understand row-wise mean using vectorized slicing and aggregation.
# calculate average of last two months
last_two_month_avg = np.mean(sales_data[:, -2:], axis=1)
# calculate average monthly growth
average_monthly_growth = (last_month_sales - first_month_sales) / (sales_data.shape[1] - 1)
#  predicting July sales
predicted_july_sales = last_two_month_avg + average_monthly_growth

print("\n--- Bonus : Predicted July sales ---")
# use zip to dispaly predicted sales for each product
for product, prediction in zip(product_names, predicted_july_sales):
    print(f"{product}: {prediction:.2f}")
````
