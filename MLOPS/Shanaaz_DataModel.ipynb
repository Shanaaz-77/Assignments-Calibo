Run the cell below to simulate **three versions** of their dataset - representing how data changed over time.

````python
# ----- DATA SIMULATOR - Do not modify ----------------------------------
np.random.seed(42)

def make_loan_dataset(n=2000, schema_version="v1", seed=42):
    """Simulate LoanGuard's loan application data at different points in time."""
    rng = np.random.RandomState(seed)
    age            = rng.randint(22, 65, n)
    income         = rng.normal(55000, 20000, n).clip(15000, 200000)
    loan_amount    = rng.normal(12000, 5000, n).clip(1000, 50000)
    credit_score   = rng.normal(680, 80, n).clip(300, 850)
    employment_yrs = rng.randint(0, 30, n)
    debt_ratio     = rng.uniform(0.05, 0.65, n)

    # Default probability driven by risk factors
    risk = (
        -0.3 * (credit_score - 680) / 80
        + 0.4 * debt_ratio
        + 0.2 * (loan_amount / income)
        - 0.1 * employment_yrs / 10
    )
    default = (1 / (1 + np.exp(-risk * 3 + rng.normal(0, 0.5, n))) > 0.5).astype(int)

    df = pd.DataFrame({
        "age": age, "annual_income": income.round(2),
        "loan_amount": loan_amount.round(2), "credit_score": credit_score.round(1),
        "employment_years": employment_yrs, "debt_to_income": debt_ratio.round(4),
        "default": default
    })

    # v2: upstream adds a new column (common real-world schema drift)
    if schema_version in ("v2", "v3"):
        df["num_accounts"] = rng.randint(1, 10, n)

    # v3: a data quality issue introduces nulls (another common real-world problem)
    if schema_version == "v3":
        null_idx = rng.choice(n, size=int(n * 0.15), replace=False)
        df.loc[null_idx, "credit_score"] = np.nan

    return df

# Generate and save three dataset versions
datasets = {}
for ver, schema, seed in [("v1", "v1", 42), ("v2", "v2", 42), ("v3", "v3", 99)]:
    df = make_loan_dataset(schema_version=schema, seed=seed)
    path = LAB_DIR / "data" / f"loans_{ver}.csv"
    df.to_csv(path, index=False)
    datasets[ver] = df
    print(f"  loans_{ver}.csv  |  {len(df):,} rows  |  {df.shape[1]} cols  |  "
          f"{df['default'].mean():.1%} default rate  |  nulls: {df.isnull().sum().sum()}")

print("\n Three dataset versions ready")
````

###  Background - How DVC Versioning Works

DVC doesn't store data in Git. It stores a **tiny pointer file** (`.dvc`) containing:
- A **checksum** (MD5 hash) of the data file
- The file path and size
- The remote storage location

When you run `dvc pull`, DVC uses the checksum to fetch the exact data file.

````python
#  BEGINNER 1C - Detect drift between dataset versions
# ----------------------------------------------------------

def compare_dataset_versions(ptr_a, ptr_b):
    """
    Compare two versioned datasets and produce a drift report.

    Returns a dict with:
      - added_cols    : columns present in b but not a
      - removed_cols  : columns present in a but not b
      - row_delta     : change in row count
      - null_delta    : change in total null count
      - drift         : dict of {col: relative_mean_shift} for drifted columns
    """
    df_a = pd.read_csv(ptr_a["path"])
    df_b = pd.read_csv(ptr_b["path"])

    cols_a = set(df_a.columns)
    cols_b = set(df_b.columns)
    
    # YOUR CODE HERE
    report = {
        "added_cols": list(cols_b - cols_a),      # columns added in new dataset
        "removed_cols": list(cols_a - cols_b),    # columns removed
        "row_delta": int(len(df_b) - len(df_a)),       # row count difference
        # difference in null values
        "null_delta": int(df_b.isnull().sum().sum() - df_a.isnull().sum().sum()),
        "drift": {}
    }

    # Detect numeric drift in shared columns
    shared_numeric = [
        c for c in cols_a & cols_b
        if df_a[c].dtype in [np.float64, np.int64]
    ]
    for col in shared_numeric:
        mean_a = df_a[col].mean()
        mean_b = df_b[col].mean()
        # YOUR CODE HERE - compute relative shift and add to report if > 5%
        # calculating relative mean shift
        if mean_a != 0:
            shift = abs(mean_b - mean_a) / abs(mean_a)

            # if shift is more than 5% we record drift
            if shift > 0.05:
                report["drift"][col] = round(shift, 4)

    return report

# ----- Compare v1 → v2, then v2 → v3 -----
print("═" * 55)
print("Comparing loans_v1  →  loans_v2")
print("═" * 55)
r12 = compare_dataset_versions(dvc_store["v1"], dvc_store["v2"])
print(json.dumps(r12, indent=2))

print("\n" + "═" * 55)
print("Comparing loans_v2  →  loans_v3  (the bad version!)")
print("═" * 55)
r23 = compare_dataset_versions(dvc_store["v2"], dvc_store["v3"])
print(json.dumps(r23, indent=2))
````

````python
# ============================================================
# ADVANCED TASK 1 - DVC-style dataset versioning system
# ============================================================

# YOUR CODE HERE

print('\n' + '='*50)
print("Task 1 - Data Versioning with DVC-Scaffolded code ")
print('='*50)
# -------------------------------------------------
# create a pointer file for a dataset (similar to DVC)
# -----------------------------------------------------
def create_dvc_pointer(path):

    path = Path(path)

    # compute md5 hash of the dataset file
    # this helps detect if the file content changes
    with open(path, "rb") as f:
        md5 = hashlib.md5(f.read()).hexdigest()

    # load the dataset to collect basic metadata
    df = pd.read_csv(path)

    pointer = {
        "path": str(path),
        "md5": md5,
        "size": path.stat().st_size,
        "rows": int(df.shape[0]),
        "cols": int(df.shape[1]),
        "columns": list(df.columns),
        "created_at": datetime.now().isoformat()
    }

 # save pointer metadata beside the dataset
    pointer_path = path.with_suffix(".dvc")
    with open(pointer_path, "w") as f:
        json.dump(pointer, f, indent=2)

    return pointer


# -----------------------------------------------------
# compare two dataset versions
# -----------------------------------------------------
def compare_dataset_versions(ptr_a, ptr_b):

    df_a = pd.read_csv(ptr_a["path"])
    df_b = pd.read_csv(ptr_b["path"])

    cols_a = set(df_a.columns)
    cols_b = set(df_b.columns)

    report = {
        "added_columns": list(cols_b - cols_a),
        "removed_columns": list(cols_a - cols_b),
        "row_diff": int(len(df_b) - len(df_a)),
        "null_diff": int(df_b.isnull().sum().sum() - df_a.isnull().sum().sum()),
        "drift": {}
    }
 # check basic statistics drift for numeric columns
    common_cols = cols_a.intersection(cols_b)

    for col in common_cols:
        if pd.api.types.is_numeric_dtype(df_a[col]):

            stats_a = {
                "mean": df_a[col].mean(),
                "std": df_a[col].std(),
                "min": df_a[col].min(),
                "max": df_a[col].max()
            }

            stats_b = {
                "mean": df_b[col].mean(),
                "std": df_b[col].std(),
                "min": df_b[col].min(),
                "max": df_b[col].max()
            }
    # analyze statistical drift for numeric columns present in both datasets
    common_cols = cols_a.intersection(cols_b)

    for col in common_cols:
        if pd.api.types.is_numeric_dtype(df_a[col]):

            stats_a = {
                "mean": df_a[col].mean(),
                "std": df_a[col].std(),
                "min": df_a[col].min(),
                "max": df_a[col].max()
            }

            stats_b = {
                "mean": df_b[col].mean(),
                "std": df_b[col].std(),
                "min": df_b[col].min(),
                "max": df_b[col].max()
            }
            col_drift = {}

            for k in stats_a:
                if stats_a[k] != 0:
                    shift = abs(stats_b[k] - stats_a[k]) / abs(stats_a[k])
                    if shift > 0.05:
                        col_drift[k] = round(float(shift), 4)

            if col_drift:
                report["drift"][col] = col_drift

    return report


# -----------------------------------------------------
# build schema from a trusted dataset
# -----------------------------------------------------
def build_golden_schema(df):

    schema = {
        "columns": list(df.columns),
        "null_tolerance": 0.05,
        "ranges": {}
    }
    # record min/max range for numeric columns
    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            schema["ranges"][col] = {
                "min": float(df[col].min()),
                "max": float(df[col].max())
            }

    return schema


# -----------------------------------------------------
# validate dataset against the schema
# -----------------------------------------------------
def validate_schema(df, schema):

    errors = []

    # check if required columns exist
    for col in schema["columns"]:
        if col not in df.columns:
            errors.append(f"Missing column: {col}")

    # check null percentage
    for col in schema["columns"]:
        if col in df.columns:
            null_rate = df[col].isnull().mean()
            if null_rate > schema["null_tolerance"]:
                errors.append(f"{col} null rate too high: {null_rate:.2f}")

    # check numeric ranges
    for col, r in schema["ranges"].items():
        if col in df.columns:
            if df[col].min() < r["min"] or df[col].max() > r["max"]:
                errors.append(f"{col} values out of expected range")

    if errors:
        raise ValueError("\n".join(errors))

    return True


# -----------------------------------------------------
# load dataset using its pointer file (like dvc pull)
# -----------------------------------------------------
def reproduce_dataset(version):

    pointer_file = LAB_DIR / "data" / f"loans_{version}.dvc"

    with open(pointer_file) as f:
        pointer = json.load(f)

    df = pd.read_csv(pointer["path"])

    # recompute checksum to verify dataset integrity
    with open(pointer["path"], "rb") as f:
        md5 = hashlib.md5(f.read()).hexdigest()

    if md5 != pointer["md5"]:
        raise ValueError("Dataset checksum mismatch")

    return df, pointer

# -----------------------------------------------------
# Demonstration
# -----------------------------------------------------

ptr_v1 = create_dvc_pointer(LAB_DIR / "data" / "loans_v1.csv")
ptr_v2 = create_dvc_pointer(LAB_DIR / "data" / "loans_v2.csv")
ptr_v3 = create_dvc_pointer(LAB_DIR / "data" / "loans_v3.csv")

print("v1 -> v2 comparison")
print(compare_dataset_versions(ptr_v1, ptr_v2))

print("\nv2 -> v3 comparison")
print(compare_dataset_versions(ptr_v2, ptr_v3))


# schema validation
df_v1 = pd.read_csv(ptr_v1["path"])
schema = build_golden_schema(df_v1)

print("\nValidating v1")
validate_schema(df_v1, schema)

print("\nValidating v3 (should fail)")
try:
    df_v3 = pd.read_csv(ptr_v3["path"])
    validate_schema(df_v3, schema)
except Exception as e:
    print(e)


# reproduce dataset
df_reproduced, pointer = reproduce_dataset("v1")

with open(pointer["path"], "rb") as f:
    check_md5 = hashlib.md5(f.read()).hexdigest()

print("\nChecksum match:", check_md5 == pointer["md5"])

            

           
````

````python
#  BEGINNER 3B - Semantic versioning
# ------------------------------------------------------------

def bump_version(current_version, bump_type):
    """
    Increment a semantic version string.

    Parameters
    ----------
    current_version : str  e.g. 'v1.2.3'
    bump_type       : str  one of 'major', 'minor', 'patch'

    Returns
    -------
    str  e.g. 'v2.0.0' for a major bump on 'v1.2.3'

    Rules
    -----
    major → increment MAJOR, reset MINOR and PATCH to 0
    minor → increment MINOR, reset PATCH to 0
    patch → increment PATCH only
    """
    # ---------------------------------------------------------------
    # Step 1: Parse current_version into major, minor, patch integers
    # ---------------------------------------------------------------
    # YOUR CODE HERE
    # remove 'v' from version and split into major, minor, patch
    # Example: "v1.2.3" -> ["1", "2", "3"]
    major, minor, patch = current_version.lstrip("v").split(".")

    # convert them from string to integers
    major = int(major)
    minor = int(minor)
    patch = int(patch)

    # ------------------------
    # Step 2: Apply bump logic
    # ------------------------
    # YOUR CODE HERE
    # apply semantic versioning rules
    if bump_type == "major":
        # major change -> increase major version and reset others
        major += 1
        minor = 0
        patch = 0

    elif bump_type == "minor":
        # minor change -> increase minor version and reset patch
        minor += 1
        patch = 0

    elif bump_type == "patch":
        # small fix -> only increment patch version
        patch += 1
        
    # construct the updated version string again
    return f"v{major}.{minor}.{patch}"


# ----- Test all bump types -----
start = "v1.0.0"
print(f"Start version      : {start}")
print(f"After patch bump   : {bump_version(start, 'patch')}   ← threshold tweak")
print(f"After minor bump   : {bump_version(start, 'minor')}   ← new training data")
print(f"After major bump   : {bump_version(start, 'major')}   ← new architecture")
print()

# Simulate LoanGuard's version history
history = [
    ("v1.0.0", "Initial release - Random Forest on loans_v1"),
    (bump_version("v1.0.0", "patch"), "Fixed decision threshold from 0.5 → 0.46"),
    (bump_version("v1.0.1", "minor"), "Retrained on loans_v2 with new num_accounts feature"),
    (bump_version("v1.1.0", "major"), "Switched to GradientBoosting - breaking output format change"),
]
print("LoanGuard model version history:")
for ver, desc in history:
    print(f"  {ver:<12} {desc}")
````

````python
# ==========================================================
# Advanced Task 4 - Model Governance System
# ==========================================================

# YOUR CODE HERE

print('\n' + '='*50)
print("Task 4 - Registry Lifecycle - Staging, Production, Archived")
print('='*50)
# setup for MLflow registry access, audit logging, and allowed stage transitions

client = MlflowClient()

AUDIT_LOG = []
AUDIT_FILE = LAB_DIR / "audit_log.json"

VALID_TRANSITIONS = {
    "None": ["Staging"],
    "Staging": ["Production"],
    "Production": ["Archived"]
}

# ----------------------------------------------------------
# log lifecycle events for governance and audit tracking
# ----------------------------------------------------------
def log_audit_event(model_name, version, from_stage, to_stage,
                    actor, reason):

    event = {
        "timestamp": datetime.now().isoformat(),
        "model_name": model_name,
        "version": version,
        "from_stage": from_stage,
        "to_stage": to_stage,
        "actor": actor,
        "reason": reason
    }

    # keep the event in memory so reports can access it quickly
    AUDIT_LOG.append(event)

    # persist the audit history so lifecycle events survive restarts
    if AUDIT_FILE.exists():
        with open(AUDIT_FILE) as f:
            existing = json.load(f)
    else:
        existing = []

    existing.append(event)

    with open(AUDIT_FILE, "w") as f:
        json.dump(existing, f, indent=2)

    return event


# ----------------------------------------------------------
# move a model version through lifecycle stages
# ----------------------------------------------------------
def transition_stage(name, version, new_stage, actor, reason):

    current = client.get_model_version(name, version)
    current_stage = current.current_stage or "None"

    # ensure only valid stage progressions are allowed
    if new_stage not in VALID_TRANSITIONS.get(current_stage, []):
        raise ValueError(f"Invalid transition {current_stage} → {new_stage}")

    # optional safety gate before promoting to production
    if new_stage == "Production":

        run = client.get_run(current.run_id)
        f1_score = run.data.metrics.get("f1", 0)

        # warn if the model quality is low, but allow the lifecycle simulation to continue
        if f1_score < 0.65:
            print("Warning: F1 below threshold, promotion normally blocked")
    
    client.transition_model_version_stage(
        name=name,
        version=version,
        stage=new_stage
    )

    log_audit_event(name, version, current_stage,
                    new_stage, actor, reason)

    print(f"{name} v{version} → {new_stage}")


# ----------------------------------------------------------
# rollback production to a previous stable version
# ----------------------------------------------------------
def rollback_production(model_name, target_version,
                        actor, reason):

    # identify the current production model
    prod = client.get_latest_versions(model_name, stages=["Production"])

    for p in prod:

        # archive the failing production model
        client.transition_model_version_stage(
            name=model_name,
            version=p.version,
            stage="Archived"
        )

        log_audit_event(model_name, p.version,
                        "Production", "Archived",
                        actor,
                        "Rollback triggered: " + reason)

    # restore the previous stable model
    client.transition_model_version_stage(
        name=model_name,
        version=target_version,
        stage="Production"
    )

    log_audit_event(model_name, target_version,
                    "Archived", "Production",
                    actor,
                    "Rollback target activated")

    print("Rollback complete → production now running", target_version)

# ----------------------------------------------------------
# governance report queries
# ----------------------------------------------------------
class GovernanceReport:

    def __init__(self, registry_name):
        self.registry_name = registry_name

    # return details of the current production model
    def current_production(self):

        prod = client.get_latest_versions(
            self.registry_name,
            stages=["Production"]
        )

        if not prod:
            return None

        version = prod[0].version

         # find audit event that promoted this version
        for e in reversed(AUDIT_LOG):
            if e["version"] == version and e["to_stage"] == "Production":
                return {
                    "name": self.registry_name,
                    "version": version,
                    "stage": "Production",
                    "promoter": e["actor"],
                    "promotion_time": e["timestamp"]
                }

    # return full lifecycle history
    def full_changelog(self, model_name):

        lines = []
        for e in AUDIT_LOG:
            if e["model_name"] == model_name:
                lines.append(
                    f"{e['timestamp']} | v{e['version']} | "
                    f"{e['from_stage']} → {e['to_stage']} | "
                    f"{e['actor']} | {e['reason']}"
                )

        return "\n".join(lines)
    # list models currently in a specific stage
    def models_in_stage(self, stage):

        versions = client.get_latest_versions(
            self.registry_name,
            stages=[stage]
        )

        return [(v.name, v.version) for v in versions]


# ==========================================================
# simulate the LoanGuard lifecycle
# ==========================================================

REGISTRY = "LoanDefaultDetector"

print("LoanGuard lifecycle simulation")

# register v1
v1 = mlflow.register_model(f"runs:/{BEST_RUN_ID}/model", REGISTRY)

transition_stage(REGISTRY, v1.version,
                 "Staging",
                 "ci-system",
                 "Offline evaluation passed")

transition_stage(REGISTRY, v1.version,
                 "Production",
                 "ml-lead@loanGuard.ai",
                 "Manual approval")

# simulate a bad model deployment
bad_model = LogisticRegression(C=0.001, max_iter=100)
bad_model.fit(X_train, y_train)

with mlflow.start_run() as bad_run:

    mlflow.log_metric("f1", 0.41)
    mlflow.sklearn.log_model(bad_model, "model")

    bad_id = bad_run.info.run_id

v2 = mlflow.register_model(f"runs:/{bad_id}/model", REGISTRY)

transition_stage(REGISTRY, v2.version,
                 "Staging",
                 "ci-system",
                 "Automated evaluation")

transition_stage(REGISTRY, v2.version,
                 "Production",
                 "auto-deploy",
                 "Bug allowed promotion")

# rollback to previous version
rollback_production(REGISTRY,
                    v1.version,
                    "ml-lead@loanGuard.ai",
                    "Production performance dropped")

# ==========================================================
# final governance report
# ==========================================================

report = GovernanceReport(REGISTRY)

print("\nCurrent Production:")
print(report.current_production())

print("\nFull Changelog:")
print(report.full_changelog(REGISTRY))

print("\nModels in Production:")
print(report.models_in_stage("Production"))

print("\nModels Archived:")
print(report.models_in_stage("Archived"))
````
