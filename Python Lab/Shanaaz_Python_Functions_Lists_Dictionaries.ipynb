```python
#----------Exercise 1.2: String Analyzer--------
print("\n--- Exercise 1.2: String Analyzer ---")
"""
Create a function analyze_string(text) that returns a dictionary with:
- "length": total number of characters
- "vowels": number of vowels (a, e, i, o, u - case insensitive)
- "consonants": number of consonants
- "words": number of words (separated by spaces)

Test with: "Python programming is amazing"
"""

# YOUR CODE HERE

#-------Exercise 1.2: String Analyzer---------

#Function to analyze string
def analyze_string(text):
  #creating dictionary
  result= {
      "length": len(text),
      "vowels": 0,
      "consonants": 0,
      "words": 0
  }
  #defining vowels and consonants
  vowels = "aeiou"
  text_lower = text.lower()

  #loop to find vowels or consonants
  for char in text_lower:
    #isalpha=check charecter in all alphabets
    if char.isalpha():
      if char in vowels:
        #number of vowel values increases in dictionary
        result["vowels"] += 1
      else:
        #number of consonents value increases in dictionay
        result["consonants"] += 1

    #number of words in text
  result["words"] = len(text.split())
  return result
# Example String
ex_str="Python programming is amazing"
print("\nGiven String:",ex_str)
#displaying result
print("\nAfter analyzing:\n")
print(analyze_string(ex_str))
```

```python

# Exercise 1.4: Grade Calculator
print("\n--- Exercise 1.4: Grade Calculator ---")
"""
Create a function calculate_grade(*scores) that:
- Accepts any number of test scores
- Returns a tuple: (average, letter_grade)
- Letter grades: A (90+), B (80-89), C (70-79), D (60-69), F (below 60)

Test with: 85, 92, 78, 90, 88
"""

# YOUR CODE HERE

#---------Exercise 1.4: Grade Calculator------

#creating function to calculate grades
def calculate_grade(*scores):
  #total score
  total= sum(scores)
  #calculating Average
  average = total / len(scores)
  #deciding letter grades
  if average >= 90:
    letter_grade = "A"
  elif average >=80: # between 80- 89
    letter_grade = "B"
  elif average >=70: # between 70 - 79
    letter_grade = "C"
  elif average >= 60: # between 60 - 69
    letter_grade = "D"
  else:                 #if average is below 60
    letter_grade = "F"
  #average and grade
  return average, letter_grade

result = calculate_grade(85, 92, 78, 90, 88)
#Displaying Result
print("\nGrades are:",result)
```

```python

# Exercise 2.2: List Statistics
print("\n--- Exercise 2.2: List Statistics ---\n")
"""
Create a function list_statistics(numbers)that returns a dictionary with:
- "sum": sum of all numbers
- "average": average of all numbers
- "minimum": smallest number
- "maximum": largest number
- "range": difference between max and min
- "count": number of elements

Test with: [15, 23, 8, 42, 16, 4, 38, 27]
"""

# YOUR CODE HERE

#----------Exercise 2.2: List Statistics---------

#creating a function that return dictionary
def list_statistics(numbers):

  total = sum(numbers) # sum of all numbers

  average = total / len(numbers) # Average of numbers

  minimum = min(numbers) # smallest number

  maximum = max(numbers) # largest number

  range = maximum - minimum # difference between max and min

  count = len(numbers) #number of elements

  #creating a dictionary
  result = {
      "sum": total,
      "average": average,
      "minimum": minimum,
      "maximum": maximum,
      "range": range,
      "count": count
  }
  return result

#  displaying result
print(list_statistics([15, 23, 8, 42, 16, 4, 38, 27]))
```

```python

# Exercise 3.2: Word Frequency Counter
print("\n--- Exercise 3.2: Word Frequency Counter ---")
"""
Create a function word_frequency(text) that:
- Takes a string as input
- Returns a dictionary with each word and its frequency
- Words should be case-insensitive
- Remove punctuation from words

Test with: "Python is great. Python is powerful. Python is easy to learn."
"""

#  YOUR CODE HERE
#----------Exercise 3.2: Word Frequency Counter---------

#creating function which returns dictionary along with word frequencies
def word_frequency(text):
  #given text
  print("\nOriginal Text:",text)

  #converting words to case-insensitive
  text = text. lower()# entire text becomes lower
  print("\nCase insensitive text:", text)

  #removing punctuations
  for char in ".,!?":
    text = text.replace(char, "") # punctuations are replacing by space
  print("\nAfter removing punctuations:",text)

  #splitting text into words
  words = text.split()
  print("\nsplitting text into words:\n",words)

  #frequency count
  #creatingan empty dictionary
  frequency = {}
  for word in words:
    if word in frequency:
      frequency[word] += 1 # if word appear more than once
    else:
      frequency[word] = 1 # word only appear once
  print("\nThe frequency of each word is:")
  return frequency

#testing by given input
input_text="Python is great. Python is powerful. Python is easy to learn."
#-------displaying result---------
print(word_frequency(input_text))
```

```python

# Exercise 3.5: Dictionary Operations
print("\n--- Exercise 3.5: Dictionary Operations ---")
"""
Create these functions:
1. merge_dictionaries(*dicts) - merges multiple dictionaries into one
2. invert_dictionary(d) - swaps keys and values

Test with:
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
dict3 = {"e": 5, "f": 6}
"""

# YOUR CODE HERE

#-------Exercise 3.5: Dictionary Operations--------

# function to merge dictionaries
def merge_dictionaries(*dicts):
  # an empty dictionary to store all other dictionaries
  merge_dict= {}
  #loop to add dictionaries
  for d in dicts:
    merge_dict.update(d)
  return merge_dict

#Creating a function to invert dictionary
def invert_dictionary(d):
  #an empty dictionary
  invert_dict = {}
  #loop to swap items
  for item in d.items():
    invert_dict[item[1]]=item[0]
  return invert_dict

#sample dictionaries
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
dict3 = {"e": 5, "f": 6}

# step 1: merging dictionaries
merged = merge_dictionaries(dict1, dict2, dict3)
print("\nstep 1: Merged dictionary:\n",merged)

# step 2: swaping keys and values
swapped= invert_dictionary(merged)
print("\nstep 2: swapped keys and values:\n",swapped)
```

```python

# ==============================================================================
# SECTION 4: INTEGRATED CHALLENGE
# ==============================================================================

print("\n" + "=" * 60)
print("SECTION 4: INTEGRATED CHALLENGE")
print("=" * 60)

# Exercise 4.1: Student Management System
print("\n--- Exercise 4.1: Student Management System ---")
"""
Create a Student Management System with these functions:

1. add_student(students_list, name, age, grades)
   - Adds student dictionary to list
   - Student dict should have: name, age, grades (list), average (calculated)

2. get_top_students(students_list, n=3)
   - Returns top n students by average grade

3. get_statistics(students_list)
   - Returns dictionary with:
     - total_students
     - overall_average
     - highest_average
     - lowest_average

4. filter_by_average(students_list, min_average)
   - Returns students with average >= min_average

Test by adding at least 5 students and using all functions.
"""

# YOUR CODE HERE

# -------Exercise 4.1: Student Management System--------

# function to add students
def add_student(students_list, name, age, grades):
  #student dictionary
  student = {
    "name": name,
    "age": age,
    "grades": grades,
    "average": sum(grades)/len(grades) #formula to calculate average
  }
  # adding student dictionary into list
  students_list.append(student)
  return students_list

# step 2: function to get top students
def get_top_students(students_list, n=3):
  #sorting list based on highest average
  sorted_students = sorted(
      students_list,
      key=lambda x : x["average"],
      reverse = True
  )
  return sorted_students[:n] # slicing upto n

#step 3: function to get statistics
def get_statistics(students_list):
  static_dict = {
      #length of list
      "total_students": len(students_list),
      #overall average among list
      "overall_average": sum(student["average"] for student in students_list) / len(students_list),
      #highest average in the list
      "highest_average": max(student["average"] for student in students_list),
      #lowest average in the list
      "lowest_average": min(student["average"] for student in students_list)
  }
  return static_dict

# step 4: Filtering by average

# function to filter by average
def filter_by_average(students_list, min_average):
  # empty list to collect filtered students
  filtered_students = []
  #loop to check students whose average is > min_average
  for student in students_list:
    if student["average"] >= min_average:
      filtered_students.append(student)
  return filtered_students

# Empty students list
students = []
# adding students
add_student(students, "Shanaaz", 20, [88,84,90])
add_student(students, "Reshma", 22, [82,68,70])
add_student(students, "Haneef", 19, [90,82,88])
add_student(students, "Prabha", 21, [85,81,82])
add_student(students, "Jyothi", 20, [70,82,72])

#--------------Displaying Result---------------

#1 Student Management System
print("\nStep 1: Student Management System")
for student in students:
  print(student)

#2 top students
print("\nStep 2: Top Studetns")
#calling get_top_students function
top_students = get_top_students(students, n=3)
#loop to get top 3 students
for student in top_students:
  print(student)

#3 to get statistics
print("\nStep 3: Statistics")
# calling get_statistics function
statistics = get_statistics(students)
print(statistics)

#4 Filtering by average
print("\nStep 4: Students with average >= minvalue")
# calling filter_by_average function
filtered_students = filter_by_average(students, 85)
for student in filtered_students:
  print(student)
```
