````python

## Exercise 5: Temperature Converter

"""**Objective**: Combine variables, input, calculations, and output.

### Requirements
Create a temperature converter that:

1. Asks for temperature in Celsius
2. Converts to Fahrenheit using: (Celsius × 9/5) + 32
3. Converts to Kelvin using: Celsius + 273.15
4. Displays all three temperatures with 2 decimal places
"""

### Write Your Code Here
# Your code here

## Exercise 5: Temperature Converter
print("\n--------Exercise 5: Temperature Converter--------\n")

# Taking input from user
celsius = float(input("Enter temperature in Celsius: "))

#conerting celsius into fahrenheit
fahrenheit = (celsius * 9/5) + 32

#converting celsius into Kelvin
kelvin = celsius + 273.15

# Displaying output
print("="*50)
print("Temperature Conversion")
print("="*50)
print(f"Celsius: {celsius:.2f}°C")
print(f"Fahrenheit: {fahrenheit:.2f}°F")
print(f"Kelvin: {kelvin:.2f}K")
````

````python

## Exercise 7: Shopping Cart Calculator

"""**Objective**: Create a practical application using all concepts learned.

### Requirements
Create a shopping cart program that:

1. Takes input for 3 products (name and price for each)
2. Calculates the subtotal
3. If subtotal is greater than $100, applies a 10% discount
4. Adds 8% tax on the discounted amount
5. Displays an itemized receipt showing:
   - Each product with price
   - Subtotal
   - Discount (if applicable)
   - Tax amount
   - Total amount

**Hint**: Use constants for tax rate (0.08), discount threshold (100), and discount rate (0.10)

### Write Your Code Here
```python"""

# Your code here

## Exercise 7: Shopping Cart Calculator
print("\n--------Exercise 7: Shopping Cart Calculator---------\n")

#Taking input for 3 products
product1_name = input("Product 1 name: ")
product1_price = float(input("Product 1 price: $"))
product2_name = input("Product 2 name: ")
product2_price = float(input("Product 2 price: $"))
product3_name = input("Product 3 name:")
product3_price = float(input("Product 3 price: $"))

print("="*50)
print("SHOPPING CART RECEIPT")
print("="*50)

# Displaying products and their prices
print(f"{product1_name:<15}${product1_price:>8.2f}")
print(f"{product2_name:<15}${product2_price:>8.2f}")
print(f"{product3_name:<15}${product3_price:>8.2f}")

print("-"*50)

# Calculating subtotal
subtotal = product1_price + product2_price + product3_price

# Constants
tax_rate = 0.08
discount_threshold = 100
discpunt_rate= 0.1

# If subtotal is greater than $100, applies a 10% discount
discount = 0
if subtotal > discount_threshold:
  discount = subtotal * discpunt_rate

# Discount amount
discount_amount = subtotal - discount

# Calculating Tax
tax = discount_amount * tax_rate

# Total amount
total = discount_amount + tax

# Displaying receipt

#{'Label':<20} -> all text starts at same column
# {value:10.2f} -> all numbers align vertically
print(f"{'Subtotal:':<20} {subtotal:>8.2f}")
print(f"{'Discount (10%):':<20} {discount:>8.2f}")
print(f"{'After Discount:':<20} {discount_amount:>8.2f}")
print(f"{'Tax (8%):':<20} {tax:>8.2f}")
print("="*50)
print(f"{'TOTAL:':<20} {total:>8.2f}")
print("="*50)
````
